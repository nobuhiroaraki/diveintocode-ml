# 【問題1】train_test_splitのスクラッチ

```python
#作成した関数
def scratch_train_test_split(X, y, train_size=0.8,random_state= None):
    """
    検証データを分割する。

    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
      訓練データ
    y : 次の形のndarray, shape (n_samples, )
      正解値
    train_size : float (0<train_size<1)
      何割をtrainとするか指定

    Returns
    ----------
    X_train : 次の形のndarray, shape (n_samples, n_features)
      訓練データ
    X_test : 次の形のndarray, shape (n_samples, n_features)
      検証データ
    y_train : 次の形のndarray, shape (n_samples, )
      訓練データの正解値
    y_test : 次の形のndarray, shape (n_samples, )
      検証データの正解値
    """
    #ここにコードを書く
    import random
    import math
    import numpy as np
    
    np.random.seed(seed=random_state)

    X_train = random.choices(X,k=math.ceil(len(X)*(train_size)))
    X_test = random.choices(X,k=math.ceil(len(X)*(1-train_size)))
    y_train =  random.choices(y,k=math.ceil(len(y)*(train_size)))
    y_test = random.choices(y,k=math.ceil(len(y)*(1-train_size)))
    
    X_train = np.array(X_train)
    X_test = np.array(X_test)
    y_train = np.array(y_train)
    y_test = np.array(y_test)
    
    return X_train, X_test, y_train, y_test
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()
X = dataset.data
y = dataset.target
```

```python
#作成した関数実行
X_train,X_test,y_train,y_test = scratch_train_test_split(X,y,random_state=1)
#結果
print("dataset:{}".format(len(X)))
print('X_train:{}'.format(len(X_train)))
print('y_test:{}'.format(len(y_test)))
print(type(X_test))

print(X_test)
```

```python
#実際の関数を使った結果
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.8,random_state=0)
print("dataset:{}".format(len(X)))
print('X_train:{}'.format(len(X_train)))
print('y_test:{}'.format(len(y_test)))
print(type(X_test))

print(X_test)
```

# 【問題2】 分類問題を解くコードの作成
## ①irisデータセット

```python
from sklearn.datasets import load_iris
dataset = load_iris()
import pandas as pd
x_df = pd.DataFrame(dataset.data,columns = ["sepal_length","sepal_width","petal_length","petal_width"])
y_df = pd.DataFrame(dataset.target,columns=["species"])
df = pd.concat([x_df,y_df],axis=1)
df = df[df["species"].isin([1,2])].reset_index(drop=True)
display(df)

X = df.iloc[:,0:4].values
y = df["species"].values
```

### A：ロジスティック回帰（SGDClassifier）

```python
X_train,X_test,y_train,y_test = scratch_train_test_split(X,y,random_state=1)

from sklearn import linear_model
clf = linear_model.SGDClassifier()
clf.fit(X_train, y_train)
```
