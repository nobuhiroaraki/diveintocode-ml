# 【問題2】フォワードプロパゲーションの実装

三層のニューラルネットワークの フォワードプロパゲーション を作成してください。

以下の説明ではノード数は1層目は400、2層目は200としますが、変更しても構いません。

各層の数式を以下に示します。

今回はそれぞれの記号が表す配列が、実装上どのようなndarrayのshapeになるかを併記してあります。

```python
batch_size = 20 # バッチサイズ
n_features = 784 # 特徴量の数
n_nodes1 = 400 # 1層目のノード数
n_nodes2 = 200 # 2層目のノード数
n_output = 10 # 出力のクラス数（3層目のノード数）
```


### 「1層目」

## $$
A_1 = X \cdot W_1 + B_1
$$

$X$ : 特徴量ベクトル (batch_size, n_features)


$W_1$ : 1層目の重み (n_features, n_nodes1)


$B_1$ : 1層目のバイアス (n_nodes1,)


$A_1$ : 出力 (batch_size, n_nodes1)

==================================================================================
### 「1層目の活性化関数」
## $$ 
Z_1 = f(A_1)
$$
$f()$ : 活性化関数


$Z_1$ 出力 (batch_size, n_nodes1)

==================================================================================


### 「2層目」
## $$
A_2 = Z_1 \cdot W_2 + B_2
$$
$W_2$ : 2層目の重み (n_nodes1, n_nodes2)


$B_2$ : 2層目のバイアス (n_nodes2,)


$A_2$ : 出力 (batch_size, n_nodes2)

==================================================================================
### 「2層目の活性化関数」
## $$
Z_2 = f(A_2)
$$
$f()$ : 活性化関数


$Z_2$ 出力 (batch_size, n_nodes2)

==================================================================================


### 「3層目(出力層)」
## $$
A_3 = Z_2 \cdot W_3 + B_3
$$
$W_3$ : 3層目の重み (n_nodes2, n_output)


$B_3$ : 3層目のバイアス (n_output,)


$A_3$ : 出力 (batch_size, n_output)

==================================================================================
### 「3層目の活性化関数」
## $$
Z_3 = softmax(A_3)
$$
$softmax()$ : ソフトマックス関数


$Z_3$ 出力 (batch_size, n_output)


$Z_3$ は各ラベル（0〜9）に対する確率の配列である。

==================================================================================

### 活性化関数（フォワードプロバゲーション）
活性化関数を作成し、フォワードプロパゲーションの中で使用します。

切り替えられるように実装することを推奨しますが、片方でも構いません。

#### 「シグモイド関数」
## $$
f(Z) = sigmoid(A) = \frac{1}{1+exp(-A)}
$$

#### 「ハイパボリックタンジェント関数」
次の数式で表されますが、np.tanhひとつで実現できます。

## $$
f(Z) = tanh(A) = \frac{exp(A) - exp(-A)}{exp(A) + exp(-A)}
$$

==================================================================================

### ソフトマックス関数
ソフトマックス関数を作成し、フォワードプロパゲーションの中で使用します。

これも活性化関数の一種ですが、多クラス分類の出力層で使われる特性上、区別して扱われることが多いです。

次の数式です。

## $$
Z_{3\_k} = \frac{exp(A_{3\_k})}{\sum_{i=1}^{n_c}exp(A_{3\_i})}
$$

$Z_{3_k}$ : $k$ 番目のクラスの確率ベクトル (batch_size,)


$A_{3_k}$ : $k$ 番目のクラスにあたる前の層からのベクトル (batch_size,)


$n_c$ : クラスの数、n_output。今回のMNISTでは10。


分母は全てのクラスに相当する値を指数関数に通した上で足し合わせたものです。

その中で、分子に $k$ 番目のクラスを持ってくることで、 $k$ 番目のクラスである確率が求まります。


これを10クラス分計算し、合わせたものが $Z_3$ です。

目的関数（損失関数）を作成します。


多クラス分類の目的関数である交差エントロピー誤差 L は次の数式です。

## $$
L = - \frac{1}{n_b}\sum_{j}^{n_b}\sum_{k}^{n_c}y_{jk} log(z_{3\_jk})
$$

$y_{ij} $: j 番目のサンプルの k 番目のクラスの正解ラベル（one-hot表現で0か1のスカラー）


$z_{3_ij} $: j 番目のサンプルの k 番目のクラスの確率（スカラー）


$n_{b} $: バッチサイズ、batch_size


$n_{c} $: クラスの数、n_output（今回のMNISTでは10）


サンプル1つあたりの誤差が求まります。

# 【問題4】バックプロパゲーションの実装
三層のニューラルネットワークのバックプロパゲーションを作成してください。確率的勾配降下法を行う部分です。


数式を以下に示します。


まず、i層目の重みとバイアスの更新式です。 $W_i$ と $B_i$ に対し、更新後の $W_i^{\prime}$ と $B_i^{\prime}$ は次の数式で求められます。



$\alpha$ : 学習率（層ごとに変えることも可能だが、基本的には全て同じとする）


$\frac{\partial L}{\partial W_i}$ : $W_i$ に関する損失 $L$ の勾配


$\frac{\partial L}{\partial B_i}$ : $B_i$ に関する損失 $L$ の勾配


＊この勾配はミニバッチのサンプル数分の合計または平均を考えます。ここでは合計を計算します。


この更新方法はSprint3線形回帰やsprint4ロジスティック回帰における最急降下法と同様です。より効果的な更新方法が知られており、それは次のSprintで扱います。


勾配 $\frac{\partial L}{\partial W_i}$ や $\frac{\partial L}{\partial B_i}$ を求めるために、バックプロパゲーションを行います。以下の数式です。ハイパボリックタンジェント関数を使用した例を載せました。シグモイド関数の場合の数式はその後ろにあります。


「3層目」



$\frac{\partial L}{\partial A_3}$ : $A_3$ に関する損失 $L$ の勾配 (batch_size, n_output)

$X$ : 特徴量ベクトル (batch_size, n_features)

## ミニバッチ処理

これまでの機械学習スクラッチでは、全てのサンプルを一度に計算していました。

しかし、ニューラルネットワークではデータを分割して入力する 確率的勾配降下法 が一般的です。

分割した際のひとかたまりを ミニバッチ 、そのサンプル数を バッチサイズ と呼びます。

今回はバッチサイズを20とします。

今回使う訓練データは48000枚ですから、48000÷20で2400回の更新を繰り返すことになります。

ニューラルネットワークではこれを2400回 イテレーション（iteration） すると呼びます。

訓練データを一度全て見ると1回の エポック（epoch） が終わったことになります。

このエポックを複数回繰り返し、学習が完了します。

これを実現するための簡素なイテレータを用意しました。for文で呼び出すと、ミニバッチを取得できます。

````python
class GetMiniBatch:
    """
    ミニバッチを取得するイテレータ

    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
      訓練データ
    y : 次の形のndarray, shape (n_samples, 1)
      正解値
    batch_size : int
      バッチサイズ
    seed : int
      NumPyの乱数のシード
    """
    def __init__(self, X, y, batch_size = 20, seed=0):
        self.batch_size = batch_size
        np.random.seed(seed)
        shuffle_index = np.random.permutation(np.arange(X.shape[0]))
        self._X = X[shuffle_index]
        self._y = y[shuffle_index]
        self._stop = np.ceil(X.shape[0]/self.batch_size).astype(np.int)
    def __len__(self):
        return self._stop
    def __getitem__(self,item):
        p0 = item*self.batch_size
        p1 = item*self.batch_size + self.batch_size
        return self._X[p0:p1], self._y[p0:p1]        
    def __iter__(self):
        self._counter = 0
        return self
    def __next__(self):
        if self._counter >= self._stop:
            raise StopIteration()
        p0 = self._counter*self.batch_size
        p1 = self._counter*self.batch_size + self.batch_size
        self._counter += 1
        return self._X[p0:p1], self._y[p0:p1]
````

# 実装したスクラッチ

````python
class ScratchSimpleNeuralNetrowkClassifier():
    """
    シンプルな三層ニューラルネットワーク分類器

    Parameters
    ----------

    Attributes
    ----------
    """
    def __init__(self,n_nodes1=400,n_nodes2=200,n_output=10,sigma=0.1,lr=0.01,epoch=10,batch_size=20,active="sigmoid",verbose = False):
        self.verbose = verbose
        self.sigma = sigma
        self.lr = lr
        self.active = active
        self.n_nodes1 = n_nodes1
        self.n_nodes2 = n_nodes2
        self.n_output = n_output
        self.epoch = epoch
        self.batch_size = batch_size
    
    #パラメータの初期値を生成する
    def default_param(self,n_input,n_nodes):
        """
        重みとバイアスの初期値を決める。

        Parameters
        ----------
        a : 入力値
        sigma : ガウス分布の標準偏差／ハイパーパラメータ 
        n_features : サンプル数#2層目は１層目のノード数、３層目は２層目のノード数
        n_nodes : n層目へのノードの数
       Returns
        -------
        default_weight : 重みの初期値／次の形のndarray, shape (n_features,n_nodes1)
        default_bias : バイアスの初期値／次の形のndarray, shape (n_nodes1,1)
        """
        default_weight = self.sigma*np.random.randn(n_input,n_nodes)
        default_bias = self.sigma*np.random.randn(1,n_nodes)
        
        return default_weight,default_bias
    
    

    #順伝播
    def forward(self,a):
        """
        Parameters
        ----------
        a : 入力値／ndarray.shape(batch_size, n_features)

       Returns
        -------
        z_3 : 各ラベルに対する確率の配列/ndarray.shape (batch_size, n_output)

        """  
        self.a_1 = a@self.w_1+self.b_1#入力値を受け取り、各層ごとに各層における重みをかけてバイアスを足す。
        self.z_1 = self.active_func(self.a_1)#そして活性化関数を通して次の層に入力値として渡す。これを全層で行う。
        #self.a_1(20, 400),self.z_1(20, 400)
        self.a_2 = self.z_1@self.w_2 + self.b_2
        self.z_2 = self.active_func(self.a_2)
        #self.a_2(20, 200),self.z_2(20, 200)
        self.a_3 = self.z_2@self.w_3 + self.b_3
        self.z_3 = self.soft_max(self.a_3)#最後の活性化関数はソフトマックス関数を通して確率ベクトルを出力する。
        #self.a_3(20, 10),self.z_3(20, 10)
        
        return self.z_3
        
    #活性化関数    
    def active_func(self,a):
        """
        Parameters
        ----------
        a : 各層に入力される値／次の形のndarray, shape (batch_size, n_nodes_n)

       Returns
        -------
        z : 各層からの出力される値／次の形のndarray, shape(batch_size, n_nodes_n)
        """

        if self.active == "sigmoid":
            z = 1/(1+np.exp(-a))

        elif self.active == "tanh":
            z = np.tanh(a)

        return z
    
    # 最終的な出力結果（確率）を返す
    def soft_max(self,final_a):
        """
        Parameters
        ----------
        final_a : 最後の層への出力／次の形のndarray, shape (batch_size, n_nodes_final)

       Returns
        -------
        z_3 : 確率ベクトル (batch_size,n_output)
        """
        z_3 = np.exp(final_a)/np.sum(np.exp(final_a),axis=1).reshape(-1,1)
        return z_3

    
    #損失関数
    def cross_entropy_error(self,y,z):
        """
        Parameters
        ----------
        y : 正解ラベル／次の形のndarray, 
        z : 
       Returns
        -------
        z : 確率(総和は１)
        """
        L = - np.sum(y * np.log(z)) / len(y)
        
        return L

    #backするときの活性化関数
    def back_active_func(self,d_z,a):
        """
        Parameters
        ----------
        d_z : 各層に出力される値（1つ前の層からの出力）／次の形のndarray, shape(batch_size, n_nodes_n)
        a : 各層に入力される値／次の形のndarray, shape (batch_size, n_nodes_n)
        """
        if self.active == "sigmoid":
            d_a = d_z*(1- self.active_func(a))*self.active_func(a)

        elif self.active == "tanh":
            d_a = d_z*(1 - np.tanh(a)**2)

        return d_a

    #誤差逆伝播
    def back(self,X,y):
        """
        Parameters
        ----------
        X : 特徴量ベクトル (batch_size, n_features)
        y : 正解ラベル(batch_size, n_output)
       Returns
        -------
        z_3 : 各ラベルに対する確率の配列/ndarray.shape (batch_size, n_output)

        """  
        ##第３層
        #第３層の出力
        self.d_a3 = self.z_3 - y#forwardでソフトマックス関数を通して算出した確率と、正解ラベルの差を求める
        #３層目のバイアス
        self.d_b3 = np.sum(self.d_a3,axis=0)#求めた差を列方向に足し合わせてバイアスを求める
        #self.d_b3.shape(1, 10)
        #３層目の重み
        self.d_w3 = self.z_2.T@self.d_a3#２層目からの出力と、求めた差の行列積を求める
        #self.d_w3.shape(200, 10)
        #２層目の出力
        self.d_z2 = self.d_a3@self.w_3.T#求めた差と求めた重みの行列積を求める

        ##第2層
        #第2層の出力
        self.d_a2 = self.back_active_func(self.d_z2,self.a_2)#２層目の出力と、重みとバイアスを加えた２層目の入力をback用の活性化関数に通す
        #2層目のバイアス
        self.d_b2 = np.sum(self.d_a2,axis=0)#求めた差を列方向に足し合わせてバイアスを求める
        #self.d_b2.shape(1, 200)
        #2層目の重み
        self.d_w2 = self.z_1.T@self.d_a2#1層目からの出力と、求めた差の行列積を求める
        #self.d_w2.shape(400, 200)
        #1層目の出力
        self.d_z1 = self.d_a2@self.w_2.T#求めた差と求めた重みの行列積を求める

        ##第1層
        #第1層の出力
        self.d_a1 = self.back_active_func(self.d_z1,self.a_1)#1層目の出力と、重みとバイアスを加えた1層目の入力をback用の活性化関数に通す
        #1層目のバイアス
        self.d_b1 = np.sum(self.d_a1,axis=0)#求めた差を列方向に足し合わせてバイアスを求める
        #self.d_b1.shape(1, 400)
        #1層目の重み
        self.d_w1 = X.T@self.d_a1#入力値と、求めた差の行列積を求める
        #self.d_w1.shape(784, 400)

    
    
    #学習（最適なパラメータを探索）
    def fit(self, X, y, X_val=None, y_val=None):
        """
        ニューラルネットワーク分類器を学習する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        X_val : 次の形のndarray, shape (n_samples, n_features)
            検証データの特徴量
        y_val : 次の形のndarray, shape (n_samples, )
            検証データの正解値
        """
        #各層における重みとバイアスの初期値を生成する
        n_input = X.shape[1]#784
        self.w_1,self.b_1 = self.default_param(n_input,self.n_nodes1)#w_1(784, 400),b_1(1,400)
        self.w_2,self.b_2 = self.default_param(self.n_nodes1,self.n_nodes2)#w_2(400,200),b_2(1,200)
        self.w_3,self.b_3 = self.default_param(self.n_nodes2,self.n_output)#w_3(200,10),b_3(1,10)
        
        

        #クロスエントロピー誤差を保存するリスト
        self.loss = []
        self.val_loss = []

        #指定したエポック分だけ繰り返す
        for i in range(self.epoch):
            
            #サンプルをミニバッチ処理して、ミニバッチずつ学習させる
            get_mini_batch = GetMiniBatch(X, y, self.batch_size,seed=i)
            
            #クロスエントロピー誤差をエポックごとに初期化
            E = 0
            
            #n回目のエポックにおけるミニバッチを取得、各ミニバッチで一連の流れを繰り返す
            for mini_X_train, mini_y_train in get_mini_batch:
                #mini_X_train.shape)(20, 784)
                #mini_y_train.shape)(20, 10)
                
                #順伝播(まず一回通る)
                self.forward(mini_X_train)#return ➡︎z_3(20,10)
                #self.a_1.shape)(20, 400)
                #self.z_1.shape)(20, 400)
                #self.a_2.shape)(20, 200)
                #self.z_2.shape)(20, 200)
                #self.a_3.shape)(20, 10)
                #self.z_3.shape)(20, 10)
                

                #誤差逆伝播(通ってきた道を引き返しながら最適な重みとバイアスを探す)
                self.back(mini_X_train,mini_y_train)#return➡︎各バイアス、重み
                #self.d_a3.shape)(20, 10)
                #self.d_b3.shape)(10,)
                #self.d_w3.shape)(200, 10)
                #self.d_a2.shape)(20, 200)
                
                #self.d_a2.shape)(20, 200)
                #self.d_b2.shape)(200,)
                #self.d_w2.shape)(400, 200)
                #self.d_a1.shape)(20, 400)
                
                #self.d_a1.shape)(20, 400)
                #self.d_b1.shape)(400,)
                #self.d_w1.shape)(784, 400)


                #パラメータを誤差逆伝播で算出したものに学習率をかけて更新（保存）
                self.w_1 -= self.d_w1*self.lr
                self.w_2 -= self.d_w2*self.lr
                self.w_3 -= self.d_w3*self.lr
                self.b_1 -= self.d_b1*self.lr
                self.b_2 -= self.d_b2*self.lr
                self.b_3 -= self.d_b3*self.lr

                #求めたパラメータで再度通る
                self.forward(mini_X_train)
                #クロスエントロピー誤差をエポックごとに保存
                E += self.cross_entropy_error(mini_y_train,self.z_3)
                
            ave_E = E/(X.shape[0]/self.batch_size)
            self.loss.append(ave_E)

            #valデータが入力された場合はvalデータも順伝播＆誤差保存
            if X_val is not None and y_val is not None: 
                self.forward(X_val)
                E += self.cross_entropy_error(y_val,self.z_3)
            ave_E = E/(X.shape[0]/self.batch_size)
            self.val_loss.append(ave_E)

            if self.verbose:
                #verboseをTrueにした際は学習過程などを出力する(各エポックごとに)
                print("Epoch{}".format(i))
                print("loss:{}".format(self.loss[-1]))
                print("val_loss:{}".format(self.val_loss[-1]))
    
    
    #推定
    def predict(self, X):
        """
        ニューラルネットワーク分類器を使い推定する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            サンプル

        Returns
        -------
            次の形のndarray, shape (n_samples, 1)
            推定結果
        """
        self.forward(X)
        true_pred = np.argmax(self.z_3,axis=1)
        return true_pred
````
