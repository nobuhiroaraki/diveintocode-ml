# 【問題1】データの取得

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris#irisデータセットを読み込む
iris_dataset = load_iris()#読み込んだデータセットを変数に

#pandasのdataframeに変換する
#dataframe・・・pandasで２次元データに対応した型

#dataをXに格納する
X = pd.DataFrame(iris_dataset.data, columns=["sepal_length","sepal_width","petal_length","petal_width"])
#DataFrameはvalues(実際のデータの値)、columns(列の名前)、index(行の名前)の三つの要素で構成されている

#target(目的変数)をyに格納する
y = pd.DataFrame(iris_dataset.target,columns=["Species"])
print(iris_dataset.target_names)#目的変数(ラベル)の名前を出力

y = y.replace({0: 'setosa', 1: "versicolor",2:"verginica"})#yの名称を置き換える
```

# 【問題2】データの結合

```python
df = pd.concat([X,y],axis=1)
#pandasのconcatメソッドを使ってaxis=1(横方向)にXとyを結合
```

### B：全ての組み合わせの散布図を一度に表示させる、散布図行列を作る

```python
sns.pairplot(df,hue="Species")
plt.show()
```

### C：４つの特徴量に関する相関係数行列を作る
#### 項目得点間の相関係数を配置した行列を、相関係数行列という。<br> 相関係数行列は対角行列であり、同じ項目同士の相関係数は1となるので、相関係数行列の対角成分はすべて1である。<br> 従って、右上か左下どちらかの非対角成分のみを報告することもある。

```python
corr_mat = df.corr(method='pearson')
print(corr_mat)
```

### D：相関係数行列のヒートマップを作る

```python
sns.heatmap(corr_mat,vmin=-1.0,vmax=1.0,center=0,annot=True,fmt='.2f',square=True,cmap="tab20")
#vmin=最小値、vmax=最大値、center=中央値、annot=Trueでマップ上に数値を表示、fmt=小数点第何位まで表示するか square=Trueで正方形に
#cmap= で色を変えるhttps://matplotlib.org/tutorials/colors/colormaps.html

plt.show()
```
