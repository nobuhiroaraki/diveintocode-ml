### B：欠損値が何パーセント含まれるかを確認する。

```python
total = df.isnull().sum(axis=0)#特徴量ごとの欠損値の合計を出す
missing_ratio = total/1460#特徴量ごとの欠損値の合計を学習データ数の合計で割って、欠損値の割合を出す
df_total = pd.DataFrame(total.sort_values(ascending=False),columns=["total"])#totalをDaraFrameとして格納、sort_valuesでソート、(ascending=False)で降順
df_missing_ratio = pd.DataFrame(missing_ratio,columns=["missing_ratio"])#missing_ratioをDaraFrameとして格納
df_missing_value = pd.concat([df_total,df_missing_ratio],axis=1)#concatで横に結合
df_missing_value.head(20)
```

# 【問題5】用語の調査
### 以下の２つの単語を知る必要があります。これらを調査してまとめてください。
https://bellcurve.jp/statistics/course/17950.html
### A：尖度

分布が正規分布からどれだけ尖っているかを表す統計量で、山の尖り度と裾の広がり度を示します。<br>正規分布より尖った分布（データが平均付近に集中し、分布の裾が重い）のときには正の値を、<br>正規分布より扁平な分布（データが平均付近から散らばり、分布の裾が軽い）のときには負の値をとります。正規分布の場合には0になります。


### B：歪度

分布が正規分布からどれだけ歪んでいるかを表す統計量で、左右対称性を示す指標のことです。<br>「右裾が長い」もしくは「右に歪んだ」もしくは「左に偏った」分布のときには正の値を、<br>「左裾が長い」もしくは「左に歪んだ」もしくは「右に偏った」分布のときには負の値をとります。<br>左右対称の分布（例えば正規分布）の場合には0になります。


# 【問題6】分布の確認
### A：seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

```python
sns.distplot(df["SalePrice"])
```

```python
#df["尖度を求めたいもの"].kurt()で尖度を求める
print("Kurtosis(尖度) :{}".format(df["SalePrice"].kurt()))
#df["歪度を求めたいもの"].skew()で歪度を求める
print("Skewness(歪度) :{}".format(df["SalePrice"].skew()))
```

### B：目的変数に対して対数変換を行う。

```python
change_log = np.log(df["SalePrice"])
print(change_log)
```

### C：対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

```python
sns.distplot(change_log)
```

```python
print("Kurtosis(尖度) :{}".format(change_log.kurt()))
print("Skewness(歪度) :{}".format(change_log.skew()))
```

### D：対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

対数変換の前後で尖度と歪度は小さくなり、なだらかな変化になった。その結果正規分布に近い分布になった。

# 【問題7】相関係数の確認

### A：相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

```python
df = df.drop("Id",axis=1)#1列目のIdは不要なので削除
corr_mat = df.corr(method='pearson')
corr_mat
```

```python
sns.heatmap(corr_mat)
```

### B：目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

```python
#相関係数行列から目的変数と相関が高い特徴量１０個を取り出す
high_corr_index = corr_mat.nlargest(11,"SalePrice")["SalePrice"].index#nlargest(a,b)でbの一番大きい数からa個取り出す。["SalePrice"]の列だけ取り出す.indexでそれらの行ラベルを取り出す。 
#全行の取り出した上位１１個のインデックスを指定して抽出、相関係数行列化
high_corr_mat = df.loc[:,high_corr_index].corr(method="pearson")     
display(high_corr_mat)

sns.heatmap(high_corr_mat,annot=True, square=True,annot_kws={'size': 10}, yticklabels=high_corr_index, xticklabels=high_corr_index)
plt.show()

"""
#別解
#相関係数行列から目的変数と相関が高い特徴量１０個を取り出す
high_corr_index = corr_mat.nlargest(11,"SalePrice")["SalePrice"].index

#high_corr_indexのshapeは(11,)であるため、ヒートマップで表示するために(11, 11)に変換する
high_corr_mat = np.corrcoef(df[high_corr_index].values.T)#np.correcoef(配列)で相関係数を求める。(df[high_corr].values)には上位１０項目＋SalePriceの元の値。Tで転置する

sns.heatmap(high_corr_mat,annot=True, square=True,annot_kws={'size': 10}, yticklabels=high_corr_index, xticklabels=high_corr_index)
plt.show()
"""
```
