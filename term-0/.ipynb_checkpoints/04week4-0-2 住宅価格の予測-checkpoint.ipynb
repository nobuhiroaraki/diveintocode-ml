```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

%matplotlib inline
pd.set_option("display.max_rows",500)
pd.set_option("display.max_columns",500)

data = pd.read_csv("/Users/nobu/Documents/データセット/home-credit-default-risk/application_train.csv")
display(data)
```

# 【問題1】練習のために特徴量の選択

```python
gla = data["GrLivArea"]
yb = data["YearBuilt"]
sp = data["SalePrice"]
```

# 【問題2】線形回帰による推定と評価
#### 線形回帰は、連続値をとる目的変数 yy と説明変数 xx（特徴量）の線形関係をモデル化します。
#### 線形関係とは、平たく言うと、説明変数が増加（減少）するのに応じて、目的変数も単調に増加（減少）する関係です。
#### 線形回帰の目的は、説明変数と目的変数の関係を表現する線形モデルの重みを学習することです。
#### 「説明変数と目的変数の散布図において、データの分布を最もよく特徴づける直線を探し出すこと」ともいえます。
#### https://qiita.com/0NE_shoT_/items/08376b08783cd554b02e

## 〜分析〜
### ①分析用にデータフレームを作成

```python
analysis_df = pd.concat([gla,yb,sp],axis=1)

analysis_df
```

### ②四分位数を確認

```python
analysis_df.describe()
```

### ③欠損値の有無を確認

```python
analysis_df.isnull().sum()
```

### ④分布を確認

```python
sns.set(palette="bright")

fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(20,6))
sns.distplot(gla,ax=ax[0])
sns.distplot(yb,ax=ax[1])
sns.distplot(sp,ax=ax[2])
plt.show()
```

### ⑤対数変換して再度分布を表示

```python
change_gla = np.log(gla)
change_yb = np.log(yb)
change_sp = np.log(sp)

fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(20,6))
sns.distplot(change_gla,ax=ax[0])
sns.distplot(change_yb,ax=ax[1])
sns.distplot(change_sp,ax=ax[2])
plt.show()
```

### ⑥対数変換したデータでDFを作り、相関係数を算出する

```python
change_df = pd.concat([change_gla,change_yb,change_sp],axis=1)
display(change_df)
display(change_df.corr(method='pearson'))
```

### ⑦目的変数と各特徴量の散布図

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(20,6))
sns.scatterplot(x="GrLivArea", y="SalePrice",data=change_df,ax=ax[0])
sns.scatterplot(x="YearBuilt", y="SalePrice",data=change_df,ax=ax[1])
plt.show()
```

## 〜前処理〜   

```python
#DFを特徴量と目的変数に分ける
X = change_df.drop("SalePrice",1)#SalePriceを削除
y = change_df["SalePrice"]
#ndarrayに変換
X = X.values
y = y.values
print(type(X))
print(type(y))
```

```python
#訓練用７５％検証用２５％に分割
X_train,X_test = train_test_split(X,test_size=0.25,random_state=1)
y_train,y_test = train_test_split(y,test_size=0.25,random_state=1)

"""
以下のエラーが出たため目的変数をreshape(-1,1)
ValueError: Expected 2D array, got 1D array instead:
Reshape your data either using array.reshape(-1, 1) if your data has a single feature or array.reshape(1, -1) if it contains a single sample. 
"""
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)

print(len(y_train))
print(len(y_test))
```

```python
#特徴量を計算・標準化(計算は訓練データのみ)
std_X = StandardScaler()

std_X.fit(X_train)
std_X_train = std_X.transform(X_train)
std_X_test = std_X.transform(X_test)

#目的変数を計算・標準化(計算は訓練データのみ)
std_y = StandardScaler()

std_y.fit(y_train)
std_y_train = std_y.transform(y_train)
std_y_test = std_y.transform(y_test)
```

## A：線形回帰
#### https://qiita.com/0NE_shoT_/items/08376b08783cd554b02e
#### https://qiita.com/yossyyossy/items/ac6961ca5a0e765f56be

## 〜学習〜

```python
#インスタンスを作成
from sklearn.linear_model import LinearRegression
LR = LinearRegression()

"""以下のエラーが出たためstd_y_trainの１の次元を削除
DataConversionWarning: 
A column-vector y was passed when a 1d array was expected. Please change the shape of y to (n_samples, ), for example using ravel().
"""
std_y_train = np.squeeze(std_y_train)
std_y_test = np.squeeze(std_y_test)

#重みを学習
LR.fit(std_X_train,std_y_train)
```

## 〜推定〜

```python
#訓練データを用いて目的変数を予測
Y_train_pred = LR.predict(std_X_train)
#検証データを用いて目的変数を予測
Y_test_pred = LR.predict(std_X_test)
```

## 〜評価（平均２乗誤差）〜

```python
from sklearn.metrics import mean_squared_error
print("訓練データ：{}".format(mean_squared_error(std_y_train,Y_train_pred)))
print("検証データ：{}".format(mean_squared_error(std_y_test,Y_test_pred)))

LR_train_MSE = mean_squared_error(std_y_train,Y_train_pred)
LR_test_MSE = mean_squared_error(std_y_test,Y_test_pred)
```

#### ※学習データ・検証データそれぞれを用いたときの平均二乗誤差を比較して、検証データを用いたときの誤差の方が大きい場合は<br>構築した線形モデルは学習データにフィットしすぎている(過学習)ことが示唆される。

### 検証データ

```python
fig,ax = plt.subplots(1,2,figsize=(20,6))

sns.scatterplot(std_X_test[:,0], std_y_test,label="std_y_test",ax=ax[0])
sns.scatterplot(std_X_test[:,0], Y_test_pred,label="Y_train_pred",ax=ax[0])
ax[0].legend(loc="upper left")
ax[0].set(xlabel ="GrLivArea",ylabel="SalePrice")

sns.scatterplot(std_X_test[:,1], std_y_test,label="std_y_test",ax=ax[1])
sns.scatterplot(std_X_test[:,1], Y_test_pred,label="Y_train_pred",ax=ax[1])
ax[1].legend(loc="upper left")
ax[1].set(xlabel ="YearBuilt",ylabel="SalePrice")
plt.show()
```

## 考察

〜MSE〜<br>
検証データにおいてMSEが最も小さかったのは線形回帰であった。<br>そのため今回の分類において適しているのは線形回帰だと考えられる。<br>
また、決定木とランダムフォレストはMSEが訓練データ＜検証データであったことから、過学習を起こしていると考えられる。<br>
しかしこれら二つはmax_depth等の調整が重要であるが、今回何も調整を行っていないため再度検証する必要がある。

〜散布図〜<br>
それぞれの手法において作成した散布図において、YearBuiltよりもGrLivAreaが直線に近い形となっていた。<br>
そのため、SalePriceとの相関の強さはYearBuilt＜GrLivAreaであると考えられる。
