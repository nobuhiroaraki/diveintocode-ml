【問題】栗まんじゅうが太陽系を覆うのにかかる時間

「バイバイン」は薬品で、これを振掛けるとその物体は5分ごとに倍に増えるという働きがあります。作中では栗まんじゅうを増やしますが、食べきれなくなったためロケットを使い宇宙に捨てました。
ここで度々話題になるのが、栗まんじゅうはその後宇宙でどのように増えていくのかについてです。様々な議論がされていますが、今回は単純に栗まんじゅうの体積がどう増えるかを計算してみましょう。
1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。
栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。
ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも適用できるような汎用的なプログラムを作成してください。また、グラフを表示するなど分かりやすいノートブックを作成してください。

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

def covered_minute(smaller,larger):#(覆うもの,覆われるもの)
    time = 0
    size_lst = []
    time_lst = []
    
    while smaller < larger:
        smaller*= 2
        time += 5
        size_lst.append(smaller)
        time_lst.append(time)
    
    return size_lst,time_lst

tmp = covered_minute(10**-4,4/3*math.pi*(15e+14)**3)
#smaller = くり饅頭１つの体積(100cm³ = 1０¯⁴m³)と仮定
#larger = 太陽系の体積(半径を15e+14 mとして計算)

result = tmp
print("答え：{}分".format(result[1][-1]))

size_lst = result[0]
solarsystem_volume = []
for i in range(180):
    solarsystem_volume.append(4/3*math.pi*(15e+14)**3)

#グラフを作成
plt.title("the count covered Solarystem by Kurimanju")
plt.xlabel("count")
plt.ylabel("volume")
plt.xlim([0,180])
plt.ylim([0,1.600000000000000e+46])
plt.plot(size_lst,color="blue")
plt.plot(solarsystem_volume,color="red")
plt.show()
```

【問題】野球ボールが東京ドームを埋め尽くすのにかかる時間

```python
tmp = covered_minute(4/3*math.pi*0.037**3,1240000)
#smaller = 野球ボール１つの体積(半径を3.7cm = 0.037m)と仮定
#larger = 東京ドームの体積(124万m³として計算)

result = tmp
print("答え：{}分".format(result[1][-1]))

size_lst = result[0]
tokyodome_volume = []
for i in range(100):
    tokyodome_volume.append(1240000)

#グラフを作成    
plt.title("the count covered TokyoDome by Baseball")
plt.xlabel("count")
plt.ylabel("volume")
plt.xlim([0,35])
plt.ylim([0,1400000])
plt.plot(size_lst,color="blue")
plt.plot(tokyodome_volume,color="red")
plt.show()
```
